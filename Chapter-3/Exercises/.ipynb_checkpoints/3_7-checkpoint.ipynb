```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread("../images/potrait2.jpeg", cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization once
equalized_1 = cv2.equalizeHist(image)

# Apply histogram equalization again on the already equalized image
equalized_2 = cv2.equalizeHist(equalized_1)

# Check if the images are identical after the second pass
is_same = np.array_equal(equalized_1, equalized_2)
print("Is the image same after second histogram equalization?", is_same)

# Function to plot images and histograms
def plot_results(original, eq1, eq2):
    fig, axes = plt.subplots(3, 2, figsize=(10, 10))

    # Original image and its histogram
    axes[0, 0].imshow(original, cmap="gray")
    axes[0, 0].set_title("Original Image")
    axes[0, 1].hist(original.ravel(), bins=256, range=[0, 256], color="black")
    axes[0, 1].set_title("Histogram (Original)")

    # First-pass equalized image and its histogram
    axes[1, 0].imshow(eq1, cmap="gray")
    axes[1, 0].set_title("After First Histogram Equalization")
    axes[1, 1].hist(eq1.ravel(), bins=256, range=[0, 256], color="black")
    axes[1, 1].set_title("Histogram (First Equalization)")

    # Second-pass equalized image and its histogram
    axes[2, 0].imshow(eq2, cmap="gray")
    axes[2, 0].set_title("After Second Histogram Equalization")
    axes[2, 1].hist(eq2.ravel(), bins=256, range=[0, 256], color="black")
    axes[2, 1].set_title("Histogram (Second Equalization)")

    plt.tight_layout()
    plt.show()

# Plot the results
plot_results(image, equalized_1, equalized_2)
```
