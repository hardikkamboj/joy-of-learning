How would the eq hist change if we set all lower order bit planes to 0?

```python
def change_value_by_bit_plane(img, bit_planes = []):
    #Iterate over each pixel and change pixel value to binary using np.binary_repr() and store it in a list.
    lst = []
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
             lst.append(np.binary_repr(img[i][j] ,width=8)) # width = no. of bits
    
    bit_planes_dict = {}
    # We have a list of strings where each string represents binary pixel value. To extract bit planes we need to iterate over the strings and store the characters corresponding to bit planes into lists.
    # Multiply with 2^(n-1) and reshape to reconstruct the bit image.
    bit_planes_dict[8] = (np.array([int(i[0]) for i in lst],dtype = np.uint8) * 128).reshape(img.shape[0],img.shape[1])
    bit_planes_dict[7] = (np.array([int(i[1]) for i in lst],dtype = np.uint8) * 64).reshape(img.shape[0],img.shape[1])
    bit_planes_dict[6] = (np.array([int(i[2]) for i in lst],dtype = np.uint8) * 32).reshape(img.shape[0],img.shape[1])
    bit_planes_dict[5] = (np.array([int(i[3]) for i in lst],dtype = np.uint8) * 16).reshape(img.shape[0],img.shape[1])
    bit_planes_dict[4] = (np.array([int(i[4]) for i in lst],dtype = np.uint8) * 8).reshape(img.shape[0],img.shape[1])
    bit_planes_dict[3] = (np.array([int(i[5]) for i in lst],dtype = np.uint8) * 4).reshape(img.shape[0],img.shape[1])
    bit_planes_dict[2] = (np.array([int(i[6]) for i in lst],dtype = np.uint8) * 2).reshape(img.shape[0],img.shape[1])
    bit_planes_dict[1] = (np.array([int(i[7]) for i in lst],dtype = np.uint8) * 1).reshape(img.shape[0],img.shape[1])
    
    for bit_plane in bit_planes:
        bit_planes_dict[bit_plane] = np.zeros_like(img)
        
    final_image = bit_planes_dict[8] + bit_planes_dict[7]+ bit_planes_dict[6]+ bit_planes_dict[5]+ bit_planes_dict[4] + bit_planes_dict[3] + bit_planes_dict[2] + bit_planes_dict[1]
    return final_image
```

```python
img_with_lower_zero = change_value_by_bit_plane(gray, bit_planes = [1,2,3,4])
```

```python
def compare_histograms(img1, img2):
    plt.figure(figsize = (20,6))
    
    plt.subplot(1,2,1)
    plt.hist(np.ravel(img1), bins=255)
    plt.title("Before removing bit plane(s)")
    
    plt.subplot(1,2,2)
    plt.hist(np.ravel(img2), bins=255)
    plt.title("After removing bit plane(s)")
    
    plt.show()
    
```

```python
def visualize_change_in_histogram(gray_img, bit_planes_to_remove = []):
    """
    Given a gray_scale image, and list of bit_planes to removes, it
    visualizes the equalized histogram of gray scale, to the image obtained by 
    removing the bit planes according to the argument bit_planes_to_remove
    """
    # get the histogram equalized image of the input image
    
    # make changes to the input image such that given bit planes are 0
    img_modified = change_value_by_bit_plane(gray_img, bit_planes=bit_planes_to_remove)
    # get the histogram equalized image for this modificatoin
    equalized_img_modified = cv2.equalizeHist(img_modified)
    
    # plot the histogram 
    compare_histograms(gray_img, equalized_img_modified)
```

```python
print("Removing bit plane - 1")
visualize_change_in_histogram(gray, [1])

print("Removing bit plane - 1, 2")
visualize_change_in_histogram(gray, [1,2])

print("Removing bit plane - 1,2,3")
visualize_change_in_histogram(gray, [1,2,3])

print("Removing bit plane - 1,2,3,4")
visualize_change_in_histogram(gray, [1,2,3,4])

```

# After removing the lower bit planes, the number of discrete values decreases, but the shape of the histogram matches the original histogram

```python
print("Removing bit plane - 8")
visualize_change_in_histogram(gray, [8])

print("Removing bit plane - 7,8")
visualize_change_in_histogram(gray, [7,8])

print("Removing bit plane - 6,7,8")
visualize_change_in_histogram(gray, [6,7,8])

print("Removing bit plane - 5,6,7,8")
visualize_change_in_histogram(gray, [5,6,7,8])

```
