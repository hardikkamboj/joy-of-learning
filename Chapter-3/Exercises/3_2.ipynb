Give a continuous function for implement- ing the contrast stretching transformation in Fig. 3.2(a). In addition to m, your function must include a parameter, E, for control- ling the slope of the function as it transi- tions from low to high intensity values. Your function should be normalized so that its minimum and maximum values are 0 and 1, respectively.

```python
import numpy as np
import matplotlib.pyplot as plt

k=0.5

# Define the contrast transformation function
def intensity_transform(r, a=10, k=0.5):
    return 1 / (1 + np.exp(-a * (r - k)))

# Generate input values
r = np.linspace(0, 1, 400)  # r ranges from 0 (dark) to 1 (light)
s = intensity_transform(r, a=10, k=0.5)

# Create the plot
plt.figure(figsize=(6, 4))
plt.plot(r, s, linewidth=2, color='b', label=r'$T(r) = \frac{1}{1 + e^{-a (r - k)}}$')

# Labels and titles
plt.xlabel(r'$r$', fontsize=12)
plt.ylabel(r'$s = T(r)$', fontsize=12)
plt.title("Sigmoid-like Intensity Transformation", fontsize=14)

# Add grid and legend
plt.grid(True, linestyle='dotted', alpha=0.7)
plt.legend()
plt.show()
```
