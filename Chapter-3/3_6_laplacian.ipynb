```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img_path = "../images/Fig0338(a)(blurry_moon).tif"

img = cv2.imread(img_path)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,8))
plt.imshow(gray, cmap="gray")
```

```python
# laplacian using cv2.filter2d 

laplacian_kernel = np.array([[0, -1, 0],
                            [-1, 4, -1], 
                            [0, -1, 0]])
```

```python
laplacian1 = cv2.filter2D(gray, -1, laplacian_kernel)
```

```python
plt.imshow(laplacian1, vmin=0, vmax=106, cmap="gray")
```

```python
# using cv2 laplacian 

laplacian2 = cv2.Laplacian(gray, -1)

plt.imshow(laplacian2, cmap="gray")
```

```python
# laplacian and normalisation

# Apply Laplacian operator
laplacian_image = cv2.Laplacian(gray, cv2.CV_64F, ksize=1)

# Convert to absolute values and scale for display
abs_laplacian = np.absolute(laplacian_image)
scaled_laplacian = np.uint8(255 * abs_laplacian / np.max(abs_laplacian))

plt.imshow(scaled_laplacian, cmap="gray")
```

```python
# enhanced image 

laplacian_image = cv2.Laplacian(gray, cv2.CV_64F, ksize=3)

# Enhance the image by subtracting the Laplacian
enhanced_image = gray - laplacian_image

# Convert to uint8 for display (clip values to avoid overflow)
enhanced_image = np.clip(enhanced_image, 0, 255).astype(np.uint8)

plt.imshow(enhanced_image, cmap="gray")
```
