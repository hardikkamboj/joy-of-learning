```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 

plt.rcParams['figure.figsize'] = [8, 6]
```

```python
img_path = "../images/Fig0340(a)(dipxe_text).tif"

img = cv2.imread(img_path)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
```

```python
# gaussian blurring 

blurred_image = cv2.GaussianBlur(gray, (31,31), 5)

plt.imshow(blurred_image, cmap="gray")
```

```python
# mask 

mask = gray.astype('float64') - blurred_image.astype('float64')

plt.imshow(mask, cmap="gray")
```

```python
unsharp_masking = cv2.addWeighted(gray.astype('float64'), 1, mask, 1, 0)

plt.imshow(unsharp_masking, cmap="gray")
```

```python
unsharp_masking = cv2.addWeighted(gray.astype('float64'), 1, mask, 4.5, 0)

plt.imshow(unsharp_masking, cmap="gray")
```

### Unsharp masking using cv2.addWeighted

```python
strength = 1.5
sharpened = cv2.addWeighted(gray, 1.0 + strength, blurred_image, -strength, 0)

plt.imshow(sharpened, cmap="gray")
```
