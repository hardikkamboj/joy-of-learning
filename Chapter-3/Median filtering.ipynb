```python
import cv2 
import matplotlib.pyplot as plt 
```

```python
img_path = "../images/Fig0335(a)(ckt_board_saltpep_prob_pt05).tif"

img = cv2.imread(img_path)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img)
```

```python
# gaussian blur 

gaussian_blur_image = cv2.GaussianBlur(gray, ksize=(19,19), sigmaX=3)

median_blur_image = cv2.medianBlur(gray, ksize=7)
```

```python
# using gray scale image

plt.figure(figsize=(25,8))

plt.subplot(131)
plt.imshow(gray, cmap="gray"); plt.title("Original Image")

plt.subplot(132)
plt.imshow(gaussian_blur_image, cmap="gray"); plt.title("Gaussian blur")

plt.subplot(133)
plt.imshow(median_blur_image, cmap="gray"); plt.title("Median blur")
```

```python
#using colored image 

gaussian_blur_image = cv2.GaussianBlur(img, ksize=(19,19), sigmaX=3)

median_blur_image = cv2.medianBlur(img, ksize=7)

plt.figure(figsize=(25,8))

plt.subplot(131)
plt.imshow(img); plt.title("Original Image")

plt.subplot(132)
plt.imshow(gaussian_blur_image); plt.title("Gaussian blur")

plt.subplot(133)
plt.imshow(median_blur_image); plt.title("Median blur")
```
