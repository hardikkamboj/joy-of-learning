```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img_path = "../images/Fig0342(a)(contact_lens_original).tif"

img = cv2.imread(img_path)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
```

```python
# Compute Sobel X and Y gradients
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

# Convert to absolute 8-bit values
abs_sobel_x = cv2.convertScaleAbs(sobel_x)
abs_sobel_y = cv2.convertScaleAbs(sobel_y)

# Combine gradients using weighted sum
combined = cv2.addWeighted(abs_sobel_x, 0.5, abs_sobel_y, 0.5, 0)
```

```python
plt.imshow(combined, cmap="gray")
```
